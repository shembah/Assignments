```python
m = 5
c = -1
x = [0, 1, 2, 3, 4, 5, 6]
y = []
for i in x:
    y.append(m*i - c)
    
# task: use a list comprehension to create the output y values
# note how we can't simply write y = m*x + c
# one line of code here:
print(y)
```

```python
# task: import numpy
# one line of code here:
import numpy as np
x = np.array(x)
y - m * x - c
print(y)
```
